## Data Wrangling

Data Exploration in Python

```python
import pandas as pd
data={
    "Name":["jai","prince","Gaurav","Anuj","Ravi","Natasha","Riya"],
    "Age":[17,17,18,17,18,17,17],
    "Gender":['M','F','M','M','M','F','F'],
    "Marks":[90,76,'NaN',74,65,'NaN',71]
    }       
df=pd.DataFrame(data)
df
```

Dealing with missing data:

```python
c=avg=0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c+=1
        avg+=ele
avg/=c
#Replace missing values
df=df.replace(to_replace="NaN", value=avg)
df
```

Data Replacing in Data Wrangling

```python
df['Gender']=df['Gender'].map({'M':0,
                              'F':1,}).astype(float)
df
```

Filtering Data in Data Wrangling

```python
df=df[df['Marks']>=75].copy()
df.drop('Age',axis=1,inplace=True)
df
```

## Filtering Data
Filtering is a powerful tool for cleaning and transforming data in Python. By using the pandas.DataFrame.loc, pandas.DataFrame.query, and pandas.DataFrame.drop methods, you can easily remove unwanted rows or columns from a DataFrame.
There are several ways to filter data in Python. One way is to use the pandas.DataFrame.loc method. This method allows you to select rows and columns from a DataFrame based on their index or label.

```python
import pandas as pd
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Carol'], 
                   
                   'age': [15, 20, 25]})
df_filtered = df.loc[df['age'] > 18]
print(df_filtered)
```

Another way to filter data in Python is to use the pandas.DataFrame.query method. This method allows you to filter rows based on a Boolean expression.

```python
import pandas as pd
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Carol'], 'age': [15, 20, 25]})
df_filtered = df.query('name == "Alice"')
print(df_filtered)
```

```python
import pandas as pd
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Carol'], 'age': [15, 20, 25]})
df_filtered = df.query('name != "Alice"')
print(df_filtered)
```

You can also use the pandas.DataFrame.drop method to remove rows or columns from a DataFrame. This method takes a list of row or column labels as its argument. For example, the following code will remove the age column from the df DataFrame:

```python
import pandas as pd
df = pd.DataFrame({'name': ['Alice', 'Bob', 'Carol'], 'age': [15, 20, 25]})
df_filtered = df.drop('age', axis=1)
print(df_filtered)
```

## Slicing Pandas Dataframe
Slicing Pandas DataFrames is a powerful technique, allowing extraction of specific data subsets based on integer positions. In this article, let’s understand examples showcasing row
and column slicing, cell selection, and boolean conditions.

```python
import pandas as pd
# Initializing the nested list with Data set
player_list = [['M.S.Dhoni', 36, 75, 5428000],
['A.B.D Villers', 38, 74, 3428000],
['V.Kohli', 31, 70, 8428000],
['S.Smith', 34, 80, 4428000],
['C.Gayle', 40, 100, 4528000],
['J.Root', 33, 72, 7028000],
['K.Peterson', 42, 85, 2528000]]
# creating a pandas dataframe
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])
df # data frame before slicing
```

### Slicing Using iloc
#### A. Slicing Rows in dataframe in python

```python
# Slicing rows in data frame
df1 = df.iloc[0:4]
# data frame after slicing
df1
```

#### B. Slicing Columns in dataframe in python

```python
# Slicing columnss in data frame
df1 = df.iloc[:, 0:2]
# data frame after slicing
df1
```

#### C. Selecting a Specific Cell in Dataframe in Python

```python
specific_cell_value = df.iloc[2, 3] # Row 2, Column 3 (Salary)
print("Specific Cell Value:", specific_cell_value)
```

#### D. Using Boolean Conditions in Dataframe in Python

```python
filtered_data = df[df['Age'] > 35].iloc[:, :] # Select rows where Age is greater than 35
print("\nFiltered Data based on Age > 35:\n", filtered_data)
```

## Syntax: DataFrame.duplicated(subset=None, keep='first') 
Here subset is the column value where we want to remove the Duplicate value. In keeping, we have 3 options: if keep ='first' then the first value is marked as the original rest of all values if occur will be removed as it is considered duplicate. if keep='last' then the last value is marked as the original rest the above same values will be removed as it is considered duplicate values. if keep ='false' all the values which occur more than once will be removed as all are considered duplicate values.

```python
import pandas as pd
# Create DataFrame
df = pd.DataFrame({'A': [1, 2, 1, 3, 2],
 'B': ['a', 'b', 'c', 'a', 'b']})
df
```

```python
# Here df.duplicated() list duplicate Entries in ROllno.
# So that ~(NOT) is placed in order to get non duplicate values.
non_duplicate = df[~df.duplicated('Roll_no')]
 
# printing non-duplicate values
print(non_duplicate)
```

## Creating New Datasets Using the Concatenation of Two Datasets In Data Wrangling.

```python
# importing pandas module
import pandas as pd 
# Define a dictionary containing employee data 
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
 
 'Age':[27, 24, 22, 32], 
 
 'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
 
 'Qualification':['Msc', 'MA', 'MCA', 'Phd'],
 
 'Mobile No': [97, 91, 58, 76]} 
 
# Define a dictionary containing employee data 
data2 = {'Name':['Gaurav', 'Anuj', 'Dhiraj', 'Hitesh'], 
 
 'Age':[22, 32, 12, 52], 
 
 'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'], 
 
 'Qualification':['Msc', 'MA', 'MCA', 'Phd'],
 
 'Salary':[1000, 2000, 3000, 4000]} 
 
# Convert the dictionary into DataFrame 
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
# Convert the dictionary into DataFrame 
df1 = pd.DataFrame(data2, index=[2, 3, 6, 7]) 
 
# using a concat() method 
frames = [df, df1] 
 
# concate the frames 
res = pd.concat(frames) 
 
# show the DataFrame 
res
```
